```python
nba['Pos'].strip()
if nba['Pos'] = 'SGa':
    nba['Pos'] = 'SG'
    
if nba['Pos'] = 'SF.':
    nba['Pos'] = 'SF'
    
if nba['Pos'] = 'PFa':
    nba['Pos'] = 'PF'

if nba['Pos'] = 'sg':
    nba['Pos'] = 'SG'

if nba['Pos'] = 'Sf':
    nba['Pos'] = 'SG'
```

```python
# Age less than 18 is not possible. Remove this row as we can't be sure of the data integrity
nba[nba['Age'] < 18]
```

```python
nba = nba[nba['Age'] >= 18]
nba = nba.reset_index(drop = True)
```

```python
# Age more than 60. Remove this row as we can't be sure of the data integrity
nba[nba['Age'] > 60]
```

```python
nba = nba[nba['Age'] <= 60]
nba = nba.reset_index(drop = True)
```

```python
# Duplicate ranked entries. These seem fine as they are the same player but different teams. Something to bear in mind when performing EDA later.
nba[nba['Rk'].duplicated()]
```

```python
# Points over 2000
nba[nba['PTS'] > 2000]
```

```python
# Points don't add up (can calculate this)
nba[3 * nba['3P'] + 2 * nba['2P'] + nba['FT'] != nba['PTS']]
```

```python
# NaN's (probably from divided by zeroes, so can probably be set to 0)
nba[nba.loc[:,'FG%':'FT%'].isnull().any(axis=1)]
```
